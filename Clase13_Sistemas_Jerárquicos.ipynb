```python
#Primer subsistema en coordenadas de Jacobi, sólo los vectores relativos

m_Jar_Aa = 1
m_Jar_Ab = 0.01

r_Jar_A = np.array([0.1, 0, 0]) #Posición relativa del cuerpo Ab respecto al centro de masa del subsistema A
v_Jar_A = np.array([0.1, 0, 0]) #Velocidad relativa del cuerpo Ab respecto al centro de masa del subsistema A
```

```python
m_Jar_Aab =  m_Jar_Aa + m_Jar_Ab   #Los dos cuerpos del subsistema A se tratan como uno solo, con la masa total del subsistema
m_Jar_B = 1.0

r_Jar_AB = np.array([1.0, 0, 0]) #Posición relativa del subsistema A respecto al subsistema B
v_Jar_AB = np.array([0, 1.0, 0]) #Velocidad relativa del subsistema A respecto al subsistema B
```

```python
#Comenzamos definiendo el centro de masa de todo el sistema

RCM_Jar_AB = np.array([0, 0, 0])    # El centro de masa de todo el sistema
VCM_Jar_AB = np.array([0, 0, 0])    # La velocidad del centro de masa de todo el sistema
```

Ahora sí podemos armar el sistema. Descendemos por el árbol del sistema jerárquico

```python
#El equivalente a la partícula 2
r_Jar_B = RCM_Jar_AB - m_Jar_Aab / (m_Jar_Aab + m_Jar_B) * r_Jar_AB 
v_Jar_B = VCM_Jar_AB - m_Jar_Aab / (m_Jar_Aab + m_Jar_B) * v_Jar_AB


# El equivalente a la partícula 1
RCM_Jar_A = RCM_Jar_AB + m_Jar_B / (m_Jar_Aab + m_Jar_B) * r_Jar_AB
VCM_Jar_A = VCM_Jar_AB + m_Jar_B / (m_Jar_Aab + m_Jar_B) * v_Jar_AB
```

En el siguiente nivel del árbol:

```python
#Partícula 1
r_Jar_Aa = RCM_Jar_A + m_Jar_Ab / (m_Jar_Aab * r_Jar_A)
v_Jar_Aa = VCM_Jar_A + m_Jar_Ab / (m_Jar_Aab * v_Jar_A)

#Partícula 2
r_Jar_Ab = RCM_Jar_A - m_Jar_Aa / (m_Jar_Aab * r_Jar_A)
v_Jar_Ab = VCM_Jar_A - m_Jar_Aa / (m_Jar_Aab * v_Jar_A)
```
