```python
import numpy as np
import pymcel as pc
```

```python
tabla, jd, X_sol = pc.consulta_horizons(
    id = '@sun',
    location = '@SSB',
    epochs = '2026-02-26 00:00:00'
)
X_sol
```

```python
tabla_jup, jd_jup, X_jup = pc.consulta_horizons(
    id = '@jupiter barycenter',
    location = '@SSB',
    epochs = '2026-02-26 00:00:00')
```

```python
r_sol_0 = X_sol[:3]
v_sol_0 = X_sol[3:]

r_jup_0 = X_jup[:3]
v_jup_0 = X_jup[3:]

r_jup_0, v_jup_0
```

```python
deltat = 1 * 86400 # 1 día en segundos
r_jup_dt = r_jup_0 + v_jup_0 * deltat
r_sol_dt = r_sol_0 + v_sol_0 * deltat

r_jup_dt, r_sol_dt
```

```python
mu_sol = pc.constantes.mu_sun
mu_jup = pc.constantes.mu_jupiter
rij_vec = r_jup_0 - r_sol_0
v_jup_dt = v_jup_0 - mu_sol * rij_vec / np.linalg.norm(rij_vec)**3 * deltat

v_jup_dt
```

```python
# aceleración sobre el Sol debida a Júpiter (vector de r del Sol hacia Júpiter = -rij_vec)
# el signo queda positivo porque rij_vec = r_jup_0 - r_sol_0
v_sol_dt = v_sol_0 + mu_jup * rij_vec / np.linalg.norm(rij_vec)**3 * deltat
v_sol_dt
```
